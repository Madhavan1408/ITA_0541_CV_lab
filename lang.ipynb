```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
llm=ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.9,
    api_key=os.getenv("GROQ_API_KEY")
)
parser=StrOutputParser()

prompt=PromptTemplate.from_template(
    """
    You are an AI Assitant  who are mastered python lanaguage 

    your goal is to analyze the code:
    1.report the bugs
    2.problems in code
    3.less compllexity code 
    4.Rewrite correct code

    dont write the unnecessary code and comments

    analyze this code :
    {"code_1"}
    
    """
)

chain = prompt | llm | parser
code="""
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser

llm=ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.9,
    api_key=os.getenv("GROQ_API_KEY")
)

parser=StrOutputParser()

classify_prompt=PromptTemplate.from_template(
    ""
    You are a sentimental analysis engine for an agriculture Platform.

    Anlayze the following farmer feedback and classify it 

    Feedback :"{feedback}",
    langauge of feedback :{language}
    crop mentioned :{crop}

    Return Only this exact format
    sentiment:[Positive/ngative/neutral],
    urgency:[High/low/medium],
    category:[price/weather,pest,logistics,other],
    recomended action :[onse sentence]
    ""
)

chain=classify_prompt | llm | parser

test_cases=[
    {
        "feedback":"Madi prices droped down by 30% this week , i am loosing money",
        "language":"engish",
        "crop":"tomato"
    },
    {
        "feedback":"The rain forecast was accurate, helped me plan my harvest",
        "language":"english",
        "crop":"rice"
    }
]

for test in test_cases:
    #print(classify_prompt.format(**test))
    print(chain.invoke(test),"\n")
    

"""

response=chain.invoke(code)
print(response)
```

# CHAT PROMPT TEMPLATE

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage,HumanMessage,AIMessage
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser
import os 
from dotenv import load_dotenv
load_dotenv()
llm =ChatGroq(model="llama-3.3-70b-versatile",temperature=0.89,top_p=0.6,api_key=os.getenv("GROQ_API_KEY"))

parser=StrOutputParser()


chat_prompt=ChatPromptTemplate(
   [
       ("system","""
       You are a Python AI devloper , an expert assistant who had experinece of teaching to students

       Rules you must follow
       - u should answervery politely to users dont irritate user
       - user should feel comfortable for using you
       - Keep Response in Normal and understandle way
       - u should teach to users in very understandle and in deep technical terms and give an experience of coding in  that concept

       """),
       ("human","{user_question}") 
    ] 
)

chain =chat_prompt | llm | parser

questions=["what is use of learning coding in age of ai",
          "do most of companies build applications on manual human type code or ai based code?",
          "What is future od coders",
          "can u guees what are the top roles that disrupted until 2036"]
for q in questions:
    response=chain.invoke(q)
    print(f"user_question : {q}","\n")
    response=chain.invoke(q)
    print(f"AI response : {response}")
    
```

```python
dynamic_system_prompt=ChatPromptTemplate.from_messages([
    ("system","""
    Yourself {bot_name}, a customer service AI for {company_name}
    Company details:
     - Industry: {industry},
     - Tone :{communication_tone},
     - Escalation email :{escalation_email}

     Critical rules:
     - Never Promise refunds without manager approval 
     - Always collect order ID before troubleshooting
     - Respond in {response_language}
    """),
    ("human","{customer_message}")
])

agrithm_config={
    "bot_name":"Agribot",
    "company_name":"Agrithm",
    "industry":"Agriculture industry",
    "communication_tone":"Friendly and simple -many users are laymans so they should feel to use it ",
    "escalation_email":"Agrithm@gmail.com",
    "response_language":"English with occasional telugu words"

}

fintech_config={
    "bot_name":"Finbot",
    "company_name":"Gelvers",
    "industry":"Financial Services",
    "communication_tone":"formal and precise",
    "escalation_email":"finance@gmail.com",
    "response_language":"English",
}

chain =dynamic_system_prompt | llm  | parser
print("--------------------------------------------------------------")
print("----------------------INTERACTIVE BOT-------------------------")
print("Enter 1 for Agriculture queries: ")
print("Enter 2 for Finance queries: ")
user_input=int(input("Enter the number for ur request: "))
querys=input("Enter ur query:  ")
if user_input==1:
    config=agrithm_config
elif user_input==2:
    config=fintech_config
else:
    print("Invalid Response")
    exit()
config['customer_message']=querys
response=chain.invoke(config)
print(response)
```
